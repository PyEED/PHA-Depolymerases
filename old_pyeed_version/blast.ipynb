## Collect the Proteins with Blast

The first step to perform further analysis with the protein family is to collect all the proteins with a blast search. You can then either add all of them to a local database or store them locally as a json file on you computer. I decided to store the results locally as a json file.    
The blast search is performed in a locally stored database which you need to provide.  
To collect all the proteins of your own family you need to change the access numbers to the one of the proteins you want to perform a blast search with. 
Furthermore, it is necessary to adapt the path to your local database in the blast search (db_path).  
After the blast search you can then choose wheter the data should be stored in a local database or on your system. For that just choose the correct code blocks. 

```python
%reload_ext autoreload
%autoreload 2
from pyeed.core import ProteinInfo
from sdrdm_database import DBConnector
import toml
import json
```

To collect the proteins 7 proteins of the 7 initial PHA Depolymerase families have been choosen to perform a blast search. The blast search was perform locally in a clustered database with less than 90% identity.   
For all searches the evalue is set to 1e-2 and n_hits adapted in order to make sure that all proteins are collect and the search is not limited by a specific number of hits. 

```python
family_1 = ProteinInfo.get_id("BAA33394.1")
blast_result_1 = family_1.blastp(
    db_path="/Users/your_path",
    evalue=1e-2, 
    n_hits=10000, 
    ncbi_key="142b47c84f365180e6838946d793b053af08")
```

I choose to write the ProteinInfo Objects I got from the blast search into json files and store them locally in a folder.  
But to perform further analysis I wanted to change the family names according to the superfamily they belong to.  

```python
for protein in blast_result_1: 
    protein.family_name = "Intracellular nPHASCL depolymerases (no lipase box)"

for protein in blast_result_2: 
    protein.family_name = "Periplasmatic PHA depolymerases"

for protein in blast_result_3:
    protein.family_name = "Intracellular nPHAMCL depolymerases"

for protein in blast_result_4:
    protein.family_name = "Extracellular dPHASCL depolymerises (catalytic domain type 1)"

for protein in blast_result_5: 
    protein.family_name = "Extracellular dPHASCL depolymerises (catalytic domain type 2)"

for protein in blast_result_6:
    protein.family_name = "Extracellular nPHASCL depolymerases"

for protein in blast_result_7: 
    protein.family_name = "Extracellular dPHAMCL depolymerase"
```
